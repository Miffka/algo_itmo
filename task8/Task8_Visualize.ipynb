```python
import os.path as osp

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(osp.join('..', 'data', 'task8.csv'))
df
```

```python
algo = 'fw'

plt.figure(figsize=(8, 6))
d = df[df['algo'] == algo].copy()
plt.scatter(d['main_n'], d['time'], label='Practical time')

coeffs = np.polyfit(x=d['main_n'] ** 3, y=d['time'], deg=1)
theor = coeffs[0] * d['main_n'] ** 3 + coeffs[1]
plt.plot(d['main_n'], theor, color='orange', label='Theoretical approximation')

plt.xlabel('Number of vertices', fontsize=16)
plt.xticks(fontsize=14)
plt.ylabel('Time, s', fontsize=16)
plt.yticks(fontsize=14)
plt.title('Floyd-Warshall algorithm time complexity', fontsize=16)
plt.legend();
```
